```python
import numpy as np
import pandas as pd
from scipy.optimize import minimize

# Generate some sample data for demonstration
np.random.seed(42)
n_samples = 100

x1 = np.array([1444, 1451, 1487, 559])
x2 = np.array([3.5, 5, 3.6, 2.7])
x3 = np.array([1, 2, 3, 4])
y1 = np.array([0.7680791 , 0.7686856 , 0.7038853 , 0.8900384 ])
y2 = np.array([0.4727152 , 0.4980309 , 0.5139198, 0.4885275 ])

# Combine the dependent variables in one array
y = np.column_stack((y1, y2))

# Combine the independent variables in one array
X = np.column_stack((x1, x2, x3))

# Log-likelihood function for MLE
def log_likelihood(params):
    a1, b1, c1, a2, b2, c2 = params
    y_pred = np.column_stack((a1 * x1 + b1 * x2 + c1 * x3, a2 * x1 + b2 * x2 + c2 * x3))
    residuals = y - y_pred
    sigma = 1.0  # Assumed constant error for simplicity
    loglik = -0.5 * np.sum(np.log(2 * np.pi * sigma**2) + (residuals**2) / sigma**2)
    return loglik

# Maximize the log-likelihood function using scipy.optimize
initial_guess = np.ones(6)  # Provide initial parameter values
result = minimize(lambda params: -log_likelihood(params), initial_guess, method='L-BFGS-B')

# Extract the estimated parameters

a1_est, b1_est, c1_est, a2_est, b2_est, c2_est = result.x

# Print the estimated parameters
print(f"a1: {a1_est}, b1: {b1_est}, c1: {c1_est},\na2: {a2_est}, b2: {b2_est}, c2: {c2_est}")
```
