```python
import pandas as pd
import numpy as np

df = pd.read_csv("/Users/pedrambeigi/Downloads/TGSIM-I695-Data.csv")
carlist = pd.read_excel("/Users/pedrambeigi/Desktop/Type.xlsx", 'CAR')
HVlist = pd.read_excel("/Users/pedrambeigi/Desktop/Type.xlsx", 'HV')
#df = df[df["time"] < 600]

lanes = [-5,-4,-3,-2]


time = df['time'].unique()
```

```python
#SPACE HEADWAY
df.sort_values(['time', 'ID'], inplace=True)
result = pd.DataFrame(columns=["lane", 'ID','time', 'IDinfront','SHeadway','Type'])

for lane in lanes:
    dfl = df[df["lane-kf"] == lane]
    
    for t in time:
        dft = dfl[dfl["time"] == t]
        
        for id1 in dft["ID"].unique():
            dfid = dft[dft['ID'] == id1]
            inputs = float(dfid['yloc-kf'])   

            df_sort = dft.iloc[(dft['yloc-kf'] - inputs).abs().argsort()[:5]]
            df_sort = df_sort[df_sort['yloc-kf'] > inputs]

            index = df_sort.index.tolist()

            if index:
                id2 = int(dft._get_value(index[0], 'ID'))
                headway = float(dft._get_value(index[0], 'yloc-kf')) - float(dfid['yloc-kf'])

                if id1 in carlist.ID.unique() and id2 in carlist.ID.unique():
                    x = 1
                elif id1 in carlist.ID.unique() and id2 in HVlist.ID.unique():
                    x = 2
                elif id2 in carlist.ID.unique() and id1 in HVlist.ID.unique():
                    x = 3
                elif id1 in HVlist.ID.unique() and id2 in HVlist.ID.unique():
                    x = 4
                else:
                    x = 0

                #print ([id1, t, int(dft._get_value(index[0], 'ID')), float(headway), str(x)])
                result.loc[len(result)] = [lane,id1, t, int(dft._get_value(index[0], 'ID')), float(headway), str(x)]
        #print (t)   
```

```python
#TIME HEADWAY
df.sort_values(['ID', 'time'], inplace=True)
result = pd.DataFrame(columns=["lane", 'ID', 'TimeinLoc340','Headway','Type',"HType"])

def has_value_between(input_list,loc):
    for value in input_list:
        if loc-10 <= value <= loc+10:
            return True
    return False


for lane in lanes:
        dfl = df[df["lane-kf"] == lane]
        
        for id1 in dfl["ID"].unique():

            y = dfl.loc[dfl['ID'] == id1, 'time'].values.tolist()
            t = dfl.loc[dfl['ID'] == id1, 'yloc-kf'].values.tolist()
            t = np.array(t)
            y = np.array(y)
            
            if has_value_between(t,610):

                poly = np.polyfit(t, y, 65)      
                poly_y = np.poly1d(poly)(610)
                
                if id1 in carlist.ID.unique():
                    x = 1
                elif id1 in HVlist.ID.unique():
                    x = 2
                else:
                    x = 0
                print ([lane, id1, poly_y, 0, x])
                result.loc[len(result)] = [lane, id1, poly_y, 0, x, 0]
```

```python
result.sort_values(['TimeinLoc340'], inplace=True)
result = result[(result.TimeinLoc340 <= 7200) & (result.TimeinLoc340 > 0)]

for lane in lanes:
    
    resultl = result[(result.lane == lane)]
    resultl = resultl.reset_index(drop=True)
    resultl['Headway'] = resultl['TimeinLoc340'].diff()
    
    for i in range (len(resultl)-1):
        
        if resultl["Type"][i] == 1 and resultl["Type"][i+1] == 1:
            x = 1
        elif resultl["Type"][i] == 1 and resultl["Type"][i+1] == 2:
            x = 2
        elif resultl["Type"][i+1] == 1 and resultl["Type"][i] == 2:
            x = 3
        elif resultl["Type"][i] == 2 and resultl["Type"][i+1] == 2:
            x = 4
        else:
            x = 0
        resultl['HType'][i] = x
    resultl.to_csv("/Users/pedrambeigi/Desktop/trb-result/Headways_610" + str(lane) + ".csv", index=False)
```
